```python
import matplotlib.pyplot as plt
import sys,os
import time
sys.path.append(os.pardir)
import simple_search
import BM
```

## 文章読み込み

```python
f = open("sample.txt","r") #38212文字空白込
tex = f.read()
```

## simple_search

```python
from simple_search import Simple_search
```

```python
simple_time = []
length = []
for i in range(1,153):
    leng = 250*i
    text = tex[:leng]
    Simple = Simple_search("Faster RCNN was not designed for pixel-to-pixel alignment between network inputs and outputs.",text)#インスタンス生成
    start = time.time()#時間計測開始
    Simple.search()
    t = time.time() - start #時間計測
    length.append(len(text)) 
    simple_time.append(t)
```

## BM_search

```python
from BM import BM_search
```

```python
bm_time = []
length2 = []
for i in range(1,153):
    leng = 250*i
    text = tex[:leng]
    BM = BM_search("Faster RCNN was not designed for pixel-to-pixel alignment between network inputs and outputs.",text)#インスタンス生成
    start = time.time()#時間計測開始
    BM.search()
    t = time.time() - start #時間計測
    length2.append(len(text)) 
    bm_time.append(t)
```

## グラフ化

```python
#1文字
p1 = plt.plot(length,simple_time)
p2 = plt.plot(length2,bm_time)
plt.xlabel("text size")
plt.ylabel("second")
plt.legend((p1[0], p2[0]), ("Simple_search", "BM_search"), loc=2)
plt.title("key:1char")
```
